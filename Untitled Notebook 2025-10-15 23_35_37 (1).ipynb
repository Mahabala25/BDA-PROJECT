```python

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Telecommunications -- Call Drop Analysis and Network Optimization") \
    .enableHiveSupport() \
    .getOrCreate()
```

```python

file_path = "/Volumes/workspace/default/newcalldata"

df_bronze = spark.read.option("header", "true") \
                      .option("inferSchema", "true") \
                      .csv(file_path)


spark.sql("CREATE SCHEMA IF NOT EXISTS bronze")


df_bronze.write.format("delta").mode("overwrite").saveAsTable("bronze.telecom_call_data")


from pyspark.sql.functions import (
    col, when, dayofweek, hour, to_timestamp,
    lag, avg, count, sum as _sum
)
from pyspark.sql.window import Window

df_silver = df_bronze.dropDuplicates()


df_silver = df_silver.withColumn("timestamp", to_timestamp(col("timestamp"), "yyyy-MM-dd HH:mm:ss"))

df_silver = df_silver.withColumn("signal_strength_dbm", col("signal_strength_dbm").cast("double")) \
                     .withColumn("tower_load_pct", col("tower_load_pct").cast("double")) \
                     .withColumn("call_duration_sec", col("call_duration_sec").cast("double")) \
                     .withColumn("handover_count", col("handover_count").cast("integer")) \
                     .withColumn("rtt_ms", col("rtt_ms").cast("double")) \
                     .withColumn("call_drop_flag", col("call_dropped").cast("integer"))


df_silver = df_silver.withColumn("weekday", dayofweek(col("timestamp")))
df_silver = df_silver.withColumn("hour", hour(col("timestamp")))
df_silver = df_silver.withColumn("is_weekend", when(col("weekday").isin([1,7]), 1).otherwise(0))


window_u = Window.partitionBy("user_id").orderBy("timestamp")
df_silver = df_silver.withColumn("prev_signal", lag("signal_strength_dbm").over(window_u))
df_silver = df_silver.withColumn("prev_rtt", lag("rtt_ms").over(window_u))

df_silver = df_silver.fillna({"prev_signal": -85.0, "prev_rtt": 100.0})

spark.sql("CREATE SCHEMA IF NOT EXISTS silver")
df_silver.write.format("delta").mode("overwrite").saveAsTable("silver.telecom_clean")


spark.sql("CREATE SCHEMA IF NOT EXISTS gold")

df_gold_tower = df_silver.groupBy("cell_tower_id").agg(
    (_sum("call_drop_flag") / count("*")).alias("DropRate"),
    avg("signal_strength_dbm").alias("AvgSignalStrength"),
    avg("tower_load_pct").alias("AvgTowerLoad"),
    avg("rtt_ms").alias("AvgRTT"),
    count("*").alias("TotalCalls")
)


df_gold_network = df_silver.groupBy("network_type").agg(
    (_sum("call_drop_flag") / count("*")).alias("DropRate"),
    avg("signal_strength_dbm").alias("AvgSignalStrength"),
    avg("call_duration_sec").alias("AvgCallDuration"),
    avg("rtt_ms").alias("AvgRTT"),
    count("*").alias("TotalCalls")
)


df_gold_tower.write.format("delta").mode("overwrite").saveAsTable("gold.tower_kpis")
df_gold_network.write.format("delta").mode("overwrite").saveAsTable("gold.network_kpis")
```

```python

from pyspark.ml.feature import VectorAssembler, StringIndexer
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import BinaryClassificationEvaluator

df_silver = df_silver.withColumn("signal_diff", col("signal_strength_dbm") - col("prev_signal"))
df_silver = df_silver.withColumn("rtt_diff", col("rtt_ms") - col("prev_rtt"))


net_indexer = StringIndexer(inputCol="network_type", outputCol="network_index")

assembler = VectorAssembler(
    inputCols=[
        "signal_strength_dbm", "tower_load_pct", "call_duration_sec",
        "handover_count", "rtt_ms", "weekday", "hour", "is_weekend",
        "prev_signal", "prev_rtt", "signal_diff", "rtt_diff", "network_index"
    ],
    outputCol="features"
)


rf = RandomForestClassifier(
    labelCol="call_drop_flag",
    featuresCol="features",
    numTrees=150,
    maxDepth=10,
    maxBins=64
)


pipeline = Pipeline(stages=[net_indexer, assembler, rf])


train_data, test_data = df_silver.randomSplit([0.7, 0.3], seed=42)


model = pipeline.fit(train_data)

preds = model.transform(test_data)

evaluator = BinaryClassificationEvaluator(labelCol="call_drop_flag")
auc = evaluator.evaluate(preds)
print(f"Random Forest AUC: {auc:.4f}")
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

df_tower = df_silver.groupBy("cell_tower_id") \
    .agg((_sum("call_drop_flag") / count("*")).alias("DropRate")) \
    .orderBy(col("DropRate").desc())

pdf_tower = df_tower.toPandas()
plt.figure(figsize=(12,6))
sns.barplot(x="DropRate", y="cell_tower_id", data=pdf_tower.head(20))
plt.title("Top 20 Cell Towers by Call Drop Rate")
plt.xlabel("Drop Rate")
plt.ylabel("Cell Tower")
plt.show()


df_signal = df_silver.groupBy("signal_strength_dbm") \
    .agg((_sum("call_drop_flag") / count("*")).alias("DropRate"))

pdf_signal = df_signal.toPandas()
plt.figure(figsize=(10,5))
sns.lineplot(x="signal_strength_dbm", y="DropRate", data=pdf_signal)
plt.title("Call Drop Rate vs Signal Strength (dBm)")
plt.xlabel("Signal Strength (dBm)")
plt.ylabel("Drop Rate")
plt.show()


df_handover = df_silver.groupBy("handover_count") \
    .agg((_sum("call_drop_flag") / count("*")).alias("DropRate"))

pdf_handover = df_handover.toPandas()
plt.figure(figsize=(10,5))
sns.barplot(x="handover_count", y="DropRate", data=pdf_handover)
plt.title("Call Drop Rate vs Handover Count")
plt.xlabel("Handover Count")
plt.ylabel("Drop Rate")
plt.show()
```
